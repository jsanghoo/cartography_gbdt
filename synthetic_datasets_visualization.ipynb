## Synthetic Datasets: visualization and noise types

A visualization of each synthetic dataset and noise types/levels is shown

1. Binary dataset:
    - first row: noise not at random (NNAR) with levels 10, 20 and 30%
    - second row: noise completely at random (NCAR) with levels 10, 20 and 30%

2. Multiclass dataset
    - first row: noise not at random (NNAR) with levels 10, 20 and 30%
    - second row: noise completely at random (NCAR) with levels 10, 20 and 30%
    
Note that the number of noisy points may seem different (especially for the noise at random) given the *density* of different classes/clusters of data.

```python
import pandas as pd
import matplotlib.pyplot as plt

nnar_synt_binary = pd.read_csv("nnar_synthethic_binary.csv")
nnar_synt_multiclass = pd.read_csv("nnar_synthethic_multiclass.csv")

ncar_synt_binary = pd.read_csv("ncar_synt.csv")
ncar_synt_multiclass = pd.read_csv("ncar_synt_multiclass.csv")
```

```python
nnar_synt_binary["colour_10"] = nnar_synt_binary.apply(lambda row: str(row["label"]) if row["label"] == row["noise_10%_nnar_coliving_neighbors_5_distance_minkowski"] else "noise", axis=1)
nnar_synt_binary["colour_20"] = nnar_synt_binary.apply(lambda row: str(row["label"]) if row["label"] == row["noise_20%_nnar_coliving_neighbors_15_distance_minkowski"] else "noise", axis=1)
nnar_synt_binary["colour_30"] = nnar_synt_binary.apply(lambda row: str(row["label"]) if row["label"] == row["noise_30%_nnar_coliving_neighbors_25_distance_minkowski"] else "noise", axis=1)

ncar_synt_binary["colour_10"] = ncar_synt_binary.apply(lambda row: str(row["label"]) if row["label"] == row["noised_label_1"] else "noise", axis=1)
ncar_synt_binary["colour_20"] = ncar_synt_binary.apply(lambda row: str(row["label"]) if row["label"] == row["noised_label_2"] else "noise", axis=1)
ncar_synt_binary["colour_30"] = ncar_synt_binary.apply(lambda row: str(row["label"]) if row["label"] == row["noised_label_3"] else "noise", axis=1)


nnar_synt_multiclass["colour_10"] = nnar_synt_multiclass.apply(lambda row: str(row["label"]) if row["label"] == row["noise_10%_nnar_coliving_neighbors_5_distance_minkowski"] else "noise", axis=1)
nnar_synt_multiclass["colour_20"] = nnar_synt_multiclass.apply(lambda row: str(row["label"]) if row["label"] == row["noise_20%_nnar_coliving_neighbors_15_distance_minkowski"] else "noise", axis=1)
nnar_synt_multiclass["colour_30"] = nnar_synt_multiclass.apply(lambda row: str(row["label"]) if row["label"] == row["noise_30%_nnar_coliving_neighbors_25_distance_minkowski"] else "noise", axis=1)

ncar_synt_multiclass["colour_10"] = ncar_synt_multiclass.apply(lambda row: str(row["label"]) if row["label"] == row["noised_label_1"] else "noise", axis=1)
ncar_synt_multiclass["colour_20"] = ncar_synt_multiclass.apply(lambda row: str(row["label"]) if row["label"] == row["noised_label_2"] else "noise", axis=1)
ncar_synt_multiclass["colour_30"] = ncar_synt_multiclass.apply(lambda row: str(row["label"]) if row["label"] == row["noised_label_3"] else "noise", axis=1)
```

```python
import seaborn as sns

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(30, 18 ))
ax = axes.flatten()

color_dict = {
    "0":'cornflowerblue',
    "1":'orange',
    'noise': 'black'
    }


alpha = 0.75


sns.scatterplot(data=nnar_synt_binary, x="feature_1", y="feature_2", hue="colour_10", palette=['cornflowerblue','orange', 'black'], ax=ax[0], alpha=alpha)
ax[0].set_title("Binary Synthetic Dataset - Noise Not At Random 10%")
ax[0].set_xlabel("Feature 1")
ax[0].set_ylabel("Feature 2")
ax[0].legend(title='Label', loc='upper right')


sns.scatterplot(data=nnar_synt_binary, x="feature_1", y="feature_2", hue="colour_20", palette=['cornflowerblue','orange', 'black'], ax=ax[1], alpha=alpha)
ax[1].set_title("Binary Synthetic Dataset - Noise Not At Random 20%")
ax[1].set_xlabel("Feature 1")
ax[1].set_ylabel("Feature 2")
ax[1].legend(title='Label', loc='upper right')

sns.scatterplot(data=nnar_synt_binary, x="feature_1", y="feature_2", hue="colour_30", palette=['cornflowerblue','orange', 'black'], ax=ax[2], alpha=alpha)
ax[2].set_title("Binary Synthetic Dataset - Noise Not At Random 30%")
ax[2].set_xlabel("Feature 1")
ax[2].set_ylabel("Feature 2")
ax[2].legend(title='Label', loc='upper right')

ncar_synt_binary.sort_values(by="colour_10", ascending=True, inplace=True)
sns.scatterplot(data=ncar_synt_binary, x="feature_1", y="feature_2", hue="colour_10", palette=color_dict, ax=ax[3], alpha=alpha)
ax[3].set_title("Binary Synthetic Dataset - Noise Completly At Random 10%")
ax[3].set_xlabel("Feature 1")
ax[3].set_ylabel("Feature 2")
ax[3].legend(title='Label', loc='upper right')

ncar_synt_binary.sort_values(by="colour_20", ascending=True, inplace=True)
sns.scatterplot(data=ncar_synt_binary, x="feature_1", y="feature_2", hue="colour_20", palette=color_dict, ax=ax[4], alpha=alpha)
ax[4].set_title("Binary Synthetic Dataset - Noise Completly At Random 20%")
ax[4].set_xlabel("Feature 1")
ax[4].set_ylabel("Feature 2")
ax[4].legend(title='Label', loc='upper right')

ncar_synt_binary.sort_values(by="colour_30", ascending=True, inplace=True)
sns.scatterplot(data=ncar_synt_binary, x="feature_1", y="feature_2", hue="colour_30", palette=color_dict, ax=ax[5], alpha=alpha)
ax[5].set_title("Binary Synthetic Dataset - Noise Completly At Random 30%")
ax[5].set_xlabel("Feature 1")
ax[5].set_ylabel("Feature 2")
ax[5].legend(title='Label', loc='upper right')



fig.savefig('binary_dataset_plots_alpha={}.png'.format(alpha), facecolor="white")
```
